```python
import psycopg2

# Establish a connection to the PostgreSQL database
connection = psycopg2.connect(
    host="localhost",
    database="dvd_rental",
    user="postgres",
    password="postgres"
)

# Create a cursor object to interact with the database
cursor = connection.cursor()

# Example SQL query to fetch some data
cursor.execute("SELECT customer_id, first_name, last_name FROM customer LIMIT 5")

# Fetch all results
results = cursor.fetchall()

# Print the fetched results
for row in results:
    print(f"Customer ID: {row[0]}, Name: {row[1]} {row[2]}")

# Close the cursor and connection
cursor.close()
connection.close()
```

```python
import pandas as pd
import psycopg2
from sqlalchemy import create_engine

# Create engine to connect to PostgreSQL
engine = create_engine('postgresql://postgres:postgres@localhost:5432/dvd_rental')

# Example SQL query to fetch data from 'customer' and 'rental'
address = pd.read_sql_query("SELECT * FROM address", engine)
category = pd.read_sql_query("SELECT * FROM category", engine)
city = pd.read_sql_query("SELECT * FROM city", engine)
country = pd.read_sql_query("SELECT * FROM country", engine)
customers = pd.read_sql_query("SELECT * FROM customer", engine)
films = pd.read_sql_query("SELECT * FROM film", engine)
film_actor = pd.read_sql_query("SELECT * FROM film_actor", engine)
film_category = pd.read_sql_query("SELECT * FROM film_category", engine)
inventory = pd.read_sql_query("SELECT * FROM inventory", engine)            
language = pd.read_sql_query("SELECT * FROM language", engine)
rentals = pd.read_sql_query("SELECT * FROM rental", engine)
payment = pd.read_sql_query("SELECT * FROM payment", engine)
staff = pd.read_sql_query("SELECT * FROM staff", engine)
store = pd.read_sql_query("SELECT * FROM store", engine)
```
